```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("./house-votes-84-updated.csv")
data.head()
```

```python
data.columns
```

```python
len(data.columns[0:])
```

```python
for i in range(0,len(data.columns)):
    data[data.columns[i]].replace("?",np.nan,inplace=True)
data.head()
```

```python
for i in range(0,len(data.columns)):
    data[data.columns[i]].replace("n",1,inplace=True)
data.head()
```

```python
for i in range(0,len(data.columns)):
    data[data.columns[i]].replace("y",2,inplace=True)
data.head()
```

```python
for i in range(0,len(data.columns)):
    data[data.columns[i]].replace("republican",3,inplace=True)
data.head()
```

```python
for i in range(0,len(data.columns)):
    data[data.columns[i]].replace("democrat",4,inplace=True)
data.head(10)
```

```python
y=data.isnull()
y.head()
```

```python
data.isnull().count()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
data.head()
data.shape
```

```python
data.info()
```

```python
from sklearn.preprocessing import Imputer
X=data.iloc[:,1:]
y=data.iloc[:,1]
#print(X)
#print(y)
##imp=Imputer(missing_value="nan",strategy="mean",axis=0)
##imp.fit(X)
##X=imp.transform(X)

imp = Imputer(missing_values = "NaN", strategy = "mean",axis=0)
imp.fit(X)
X = imp.transform(X)

new_df = pd.DataFrame(X[:10])
print(new_df)
```

```python
from sklearn.preprocessing import Imputer
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
```

```python
steps=[("imp",Imputer(missing_values="NaN",strategy="mean",axis=0)),
      ("clf",KNeighborsClassifier(n_neighbors=6))]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.3,random_state=42)

pipeline=Pipeline(steps)
pipeline.fit(X_train, y_train)
pipeline.predict(X_test)
pipeline.score(X_test, y_test)
```
