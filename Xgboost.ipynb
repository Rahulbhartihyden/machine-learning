```python
import xgboost as xgb
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
kd=pd.read_csv("./chronic_kidney_with_class.csv")
kd.head()
```

```python
kd.replace("normal",1,inplace=True)
kd.replace("abnormal",0,inplace=True)
kd.replace("notpresent",0,inplace=True)
kd.replace("present",1,inplace=True)
kd.replace("yes",1,inplace=True)
kd.replace("no",0,inplace=True)
kd.replace("good",1,inplace=True)
kd.replace("poor",0,inplace=True)
kd.replace("?",np.nan,inplace=True)
kd.replace("ckd",1,inplace=True)
kd.replace("notckd",0,inplace=True)
```

```python
kd.drop("Unnamed: 0",axis=1,inplace=True)
kd.head()
```

```python
kd.drop("dm",axis=1,inplace=True)
```

```python
from sklearn.preprocessing import Imputer
```

```python
imp=Imputer(missing_values="NaN",strategy="mean",axis=0)
```

```python
print(imp)
```

```python
X=kd.iloc[:,:-1]
y=kd.iloc[:,-1]
imp.fit(X)
X=imp.transform(X)
```

```python
# Create the training and test sets
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=123)

# Instantiate the XGBClassifier: xg_cl
xg_cl = xgb.XGBClassifier(objective='binary:logistic', n_estimators=10, seed=123)

# Fit the classifier to the training set
xg_cl.fit(X_train,y_train)

# Predict the labels of the test set: preds
preds = xg_cl.predict(X_test)

# Compute the accuracy: accuracy
accuracy = float(np.sum(preds==y_test))/y_test.shape[0]
print("accuracy:  ", (accuracy))
```
